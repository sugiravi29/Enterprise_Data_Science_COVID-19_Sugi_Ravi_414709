# Relational Data Structure

```python
data_path='../data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
pd_raw=pd.read_csv(data_path)
pd_raw.head()
```

```python
pd_data_base=pd_raw.rename(columns={'Country/Region':'country','Province/State':'state'})
pd_data_base['state']=pd_data_base['state'].fillna('no')
pd_data_base=pd_data_base.drop(columns={'Lat','Long'},axis=1)
pd_data_base.head()
```

```python
pd_relational_model=pd_data_base.set_index(['state','country']) \
                             .T                     \
                             .stack(level=[0,1])    \
                             .reset_index()         \
                             .rename(columns={'level_0':'date',
                                            0:'confirmed'},)
pd_relational_model.head()
```

```python
pd_relational_model['date']=pd_relational_model.date.astype('datetime64[ns]')
pd_relational_model.dtypes
```

```python
pd_relational_model.confirmed=pd_relational_model.confirmed.astype(int)
pd_relational_model.to_csv('../data/processed/COVID_relational_confirmed.csv',sep=';',index=False)
```

```python
print(pd_relational_model[pd_relational_model['country']=='US'].tail())
```

```python
pd_JH_data=pd.read_csv('../data/processed/COVID_relational_confirmed.csv',sep=';',parse_dates=[0])
pd_JH_data=pd_JH_data.sort_values('date',ascending=True).reset_index(drop=True).copy()
pd_JH_data.head()
```

# New structure of the Johns Hopkins data set the state information US is in a second file

```python
data_path='../data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
pd_raw_US=pd.read_csv(data_path)
pd_raw_US.head()
```

```python
pd_raw_US=pd_raw_US.drop(['UID','iso2','iso3','code3','FIPS','Admin2','Country_Region','Lat','Long_','Combined_Key'],axis=1)
pd_data_base_US=pd_raw_US.rename(columns={'Province_State':'state'}).copy()
pd_data_base_US
```

```python
pd_relational_model_US=pd_data_base_US.set_index(['state']) \
                             .T                             \
                             .stack()                       \
                             .reset_index()                 \
                             .rename(columns={'level_0':'date',
                                            0:'confirmed'})
pd_relational_model_US['country']='US'
pd_relational_model_US['date']=[datetime.strptime( each,"%m/%d/%y") for each in pd_relational_model_US.date] # convert to datetime
pd_relational_model_US.head()
```

```python
pd_relational_model_all=pd_relational_model[pd_relational_model['country']!='US'].reset_index(drop=True)
```

```python
pd_relational_model_all=pd.concat([pd_relational_model_all,pd_relational_model_US],ignore_index=True,sort=True)
```

```python
cols=list(pd_relational_model_all)
cols
```

```python
cols.insert(0, cols.pop(cols.index('date')))
cols.insert(1, cols.pop(cols.index('state')))
cols.insert(2, cols.pop(cols.index('country')))
cols.insert(3, cols.pop(cols.index('confirmed')))

pd_relational_model_all = pd_relational_model_all.loc[:, cols]
pd_relational_model_all
```

```python
pd_relational_model_all.to_csv('../data/processed/20200824_COVID_relational_confirmed.csv',sep=';',index=False)
```
